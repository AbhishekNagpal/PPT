```python
def calculate_mode(li):
    d = {}
    for i in li:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
            
    mx = li[0]
    for j in d.values():
        mx = max(mx,j)
        
    return mx
li = [4, 5, 2, 3, 5, 4, 3, 2, 4, 5]
print(calculate_mode(li))
```

```python
def calculate_median(arr):
    n = len(arr)
    arr.sort()
 
    if n % 2 == 0:
        median1 = arr[n//2]
        median2 = arr[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = arr[n//2]
        
    return median

A = [5000, 6000, 5500, 7000]
B = [4500, 5500, 5800, 6000, 5200]
print(calculate_median(A))
print(calculate_median(B))
```

```python
9. Scenario: A study is conducted to compare the effectiveness of two different medications. 
    The recovery times of the patients in each group are as follows:
   Medication A: [10, 12, 14, 11, 13]
   Medication B: [15, 17, 16, 14, 18]
   Perform an analysis of variance (ANOVA) to determine if there is a significant difference 
in the mean recovery times between the two medications.
```

```python
Step 1: Calculate the means of each group.
Mean of Medication A (x̄A) = (10 + 12 + 14 + 11 + 13) / 5
= 60 / 5
= 12

Mean of Medication B (x̄B) = (15 + 17 + 16 + 14 + 18) / 5
= 80 / 5
= 16

Step 2: Calculate the sum of squares within groups (SSW).
SSW = Σ(xi - x̄i)^2
= (10 - 12)^2 + (12 - 12)^2 + (14 - 12)^2 + (11 - 12)^2 + (13 - 12)^2
+ (15 - 16)^2 + (17 - 16)^2 + (16 - 16)^2 + (14 - 16)^2 + (18 - 16)^2
= 2^2 + 0^2 + 2^2 + 1^2 + 1^2 + 1^2 + 1^2 + 0^2 + 2^2 + 2^2
= 4 + 0 + 4 + 1 + 1 + 1 + 1 + 0 + 4 + 4
= 20

Step 3: Calculate the sum of squares between groups (SSB).
SSB = nA * (x̄A - x̄)^2 + nB * (x̄B - x̄)^2
= 5 * (12 - 14)^2 + 5 * (16 - 14)^2
= 5 * (-2)^2 + 5 * 2^2
= 5 * 4 + 5 * 4
= 20 + 20
= 40

Step 4: Calculate the degrees of freedom.
Degrees of freedom between groups (dfB) = number of groups - 1
= 2 - 1
= 1

Degrees of freedom within groups (dfW) = total number of observations - number of groups
= 10 - 2
= 8

Step 5: Calculate the mean squares (MSB and MSW).
MSB = SSB / dfB
= 40 / 1
= 40

MSW = SSW / dfW
= 20 / 8
= 2.5

Step 6: Calculate the F-value.
F = MSB / MSW
= 40 / 2.5
= 16

Step 7: Determine the critical F-value for the desired significance level.
For a significance level of α = 0.05 and dfB = 1, dfW = 8, the critical F-value is approximately 5.32 (obtained from an F-table or statistical software).

Step 8: Compare the calculated F-value with the critical F-value.
Since the calculated F-value (16) is greater than the critical F-value (5.32), we reject the null hypothesis.

Step 9: Interpret the result.
Based on the ANOVA test, there is sufficient evidence to conclude that there is a significant difference in the mean recovery times between Medication A and Medication B.
```

```python
15. Scenario: A study is conducted to compare the preferences of two groups of participants. 
    The preferences are measured on a Likert scale from 1 to 5. The data collected is as 
    follows:
    Group A: [4, 3, 5, 2, 4]
    Group B: [3, 2, 4, 3, 3]
    Perform a Mann-Whitney U test to determine if there is a significant difference in the 
    median preferences between the two groups.
```

```python
Step 1: Combine the data from both groups and assign ranks to each observation.
Combined data: [4, 3, 5, 2, 4, 3, 2, 4, 3, 3]
Ranks: [7, 4.5, 10, 1, 7, 4.5, 1, 7, 4.5, 4.5]

Step 2: Calculate the sum of ranks for each group.
Sum of ranks for Group A = 7 + 4.5 + 10 + 1 + 7
= 29.5

Sum of ranks for Group B = 4.5 + 1 + 7 + 4.5 + 4.5
= 22

Step 3: Calculate the U statistic for each group.
U statistic for Group A = nA * nB + (nA * (nA + 1) / 2) - sum of ranks for Group A
= 5 * 5 + (5 * (5 + 1) / 2) - 29.5
= 25 + (5 * 6 / 2) - 29.5
= 25 + (30 / 2) - 29.5
= 25 + 15 - 29.5
= 40 - 29.5
= 10.5

U statistic for Group B = nA * nB + (nB * (nB + 1) / 2) - sum of ranks for Group B
= 5 * 5 + (5 * (5 + 1) / 2) - 22
= 25 + (5 * 6 / 2) - 22
= 25 + (30 / 2) - 22
= 25 + 15 - 22
= 40 - 22
= 18

Step 4: Determine the minimum U statistic.
Minimum U statistic = min(U statistic for Group A, U statistic for Group B)
= min(10.5, 18)
= 10.5

Step 5: Calculate the expected value of the minimum U statistic under the null hypothesis.
Expected value = (nA * nB) / 2
= (5 * 5) / 2
= 12.5

Step 6: Calculate the standard deviation of the minimum U statistic under the null hypothesis.
Standard deviation = sqrt((nA * nB * (nA + nB + 1)) / 12)
= sqrt((5 * 5 * (5 + 5 + 1)) / 12)
= sqrt((125) / 12)
= sqrt(10.4167)
= 3.225

Step 7: Calculate the z-score.
z-score = (minimum U statistic - expected value) / standard deviation
= (10.5 - 12.5) / 3.225
= -2 / 3.225
= -0.620

Step 8: Determine the critical z-value for the desired significance level.
For a significance level of α = 0.05, the critical z-value is approximately -1.96 (obtained from a standard normal distribution table or statistical software).

Step 9: Compare the calculated z-score with the critical z-value.
Since the calculated z-score (-0.620) is greater than the critical z-value (-1.96), we fail to reject the null hypothesis.

Step 10: Interpret the result.
Based on the Mann-Whitney U test, there is insufficient evidence to conclude that there is a significant difference in the median preferences between Group A and Group B.
```
